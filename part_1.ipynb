```python
import numpy as np
import keras
```

Первая строка определяет саму модель. Модель — это обучаемая нейронная сеть, в данном случае простейшая нейронная сеть, которая только бывает.
`model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])`

Эта сеть стоить всего из одного условия. Это задается вызовом `keras.layers.Dense(units=1, input_shape=[1])` В єтом условие находится всего 1 нейрон это задано параметром `units=1`
Также мы подаем единственное значение в нейронную сеть значение переменной x.
И мы предлагаем нашей нейронной сети предсказать, чему равно y для данного x.
Поэтому ми задаем значение параметра `input_shape=[1]`.

`model.compile(optimizer='sgd', loss='mean_squared_error')` Для подготовки модели нам необходимо передать 2 функции: Функции потерь и оптимизации. Это ключевее функции в машинном обучении, суть которого состоит в том, что модель пытается угадать зависимость между переменными.

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=499)
```

```python
print(model.predict([10.0]))

# import tensorflow as tf
# print(tf.__version__)
```
